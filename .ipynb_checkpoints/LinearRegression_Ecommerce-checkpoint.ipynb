# Problem Statement:

Congratulations! You just got some contract work with an Ecommerce company based in New York City that sells clothing online but they also have in-store style and clothing advice sessions. Customers come in to the store, have sessions/meetings with a personal stylist, then they can go home and order either on a mobile app or website for the clothes they want.

The company is trying to decide whether to focus their efforts on their mobile app experience or their website. They've hired you on contract to help them figure it out! Let's get started!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Get Data

```python
ecom_data = pd.read_csv("Ecommerce Customers")
```

```python
ecom_data.head()
```

```python
ecom_data.info()
```

```python
ecom_data.describe()
```

# Exploratory Data Analysis

 jointplot to compare the Time on Website and Yearly Amount Spent columns. Does the correlation make sense

```python
sns.set_style('whitegrid')
sns.jointplot(x='Time on Website',y='Yearly Amount Spent',data=ecom_data)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.jointplot(x='Time on App',y='Yearly Amount Spent',data=ecom_data,kind='hex')
plt.show()
```

relationships across the entire data set

```python
sns.pairplot(ecom_data)
plt.show()
```

From the above graph, Length of Membership is most correlated feature with Yearly Amount Spent.
To check this relationship in more detail, lets create a linear Model plot

```python
sns.lmplot(x='Length of Membership',y='Yearly Amount Spent',data=ecom_data)
plt.show()
```

Creating Train & Test Data

```python
ecom_data.columns
```

```python
y = ecom_data['Yearly Amount Spent']
```

```python
X = ecom_data[['Length of Membership','Time on App','Time on Website','Avg. Session Length']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state = 101)
```

# Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

# Model Evaluation

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

# Predicting Test Data

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y test')
plt.ylabel('Predictions')
plt.show()
```

# Evaluating the Model

Calculate MSE, MAE, RMSE

```python
from sklearn import metrics
```

```python
print("MAE:",metrics.mean_absolute_error(y_test,predictions))
print("MSE:",metrics.mean_squared_error(y_test,predictions))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
print("R-square:", metrics.explained_variance_score(y_test,predictions))
```

- R-square explains the model variance
- Therefore 99% explains a very a good model fit

# Residuals

Explore the residuals to make sure everything was okay with our data.

```python
sns.distplot((y_test-predictions),bins=50);
```

- From the above graph, we can say that data is normally distributed

# Conclusion

```python
coeff_df
```

- From the above Coefficient table, we can say holding all other features fixed, a 1 unit increase in Time on Website is associated with an increase of 0.19 total dollars spent.
- While holding all other features fixed, a 1 unit increase in Time on App is associated with an increase of 38.59 total dollars spent.

- By looking at correlation coefficient, 
Solution 1 - Company should focus more on their mobile app which will lead to an increase in the company revenue.
Solution 2 - Develop the Website to catch up to the performance of the mobile app.
